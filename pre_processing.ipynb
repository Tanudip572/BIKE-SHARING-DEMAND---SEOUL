# **Feature Selection and Engineering**

# Feature: *Hour*
$\textit{Hour}$ is the Hour of the day. So we should treat it as a categorical feature.

```python
df['Hour']=df['Hour'].astype(str)
```

# Feature: *Functioning Day*

```python
df.groupby('Functioning Day')['Rented Bike Count'].sum().sort_values(ascending=False).reset_index()
```

So bikes have been rented only on a functioning day. So we remove rows with non-functioning days and then remove $\textit{Functioning Day}$ also.


```python
print((df['Functioning Day']=='No').sum())
```

```python
df.drop(df[df['Functioning Day']=='No'].index, inplace=True)
df.drop('Functioning Day', axis=1, inplace=True)
```

```python
df.shape
```

# Feature: *Date*

```python
df['Date']=pd.to_datetime(df['Date'],dayfirst=True)

min_date=df['Date'].min()
max_date=df['Date'].max()

print("Minimum Date: ",min_date)
print("Maximum Date: ",max_date)
```

```python
#extracting Day, Month, Year and Weekday from Date column

df['Day']=df['Date'].dt.day.astype('category')
df['Month']=df['Date'].dt.month.astype('category')
df['Year']=df['Date'].dt.year.astype('category')
df['Weekday']=df['Date'].dt.day_name().astype('category')
df.head()
```

```python
#creating a weekend indicator
df['Weekend'] = (((df['Weekday'] == 'Saturday') | (df['Weekday'] == 'Sunday')).astype('int')).astype('category')
```

```python
df.drop(['Weekday'], axis = 1, inplace=True)
```

```python
#creating a weekend or holiday indicator
df['Weekend_or_Holiday'] = ((df['Weekend'] == 1) | (df['Holiday'] == 'Yes')).astype('int').astype('category')
```

```python
df.drop(['Holiday', 'Weekend'], axis = 1, inplace=True)
```

# Feature: *Seasons*

```python
df['Seasons']=df['Seasons'].astype('category')
```
